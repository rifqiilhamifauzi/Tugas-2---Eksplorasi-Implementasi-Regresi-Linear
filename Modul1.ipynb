Modul 1: Menggunakan data sembarang dalam
mengimplementasikan model Regresi Linear

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Membuat data sembarang
penjualan = np.array([6,5,5,4,4,3,2,2,2,1])
harga = np.array([16000, 18000, 27000, 34000, 50000, 68000, 65000, 81000, 85000, 90000])

# Visualisasi awal
plt.scatter(penjualan, harga)
plt.xlabel("Penjualan")
plt.ylabel("Harga")
plt.title("Grafik data penjualan dan harga")
plt.show()
```

```python
# Reshape data agar berbentuk matriks kolom[cite: 1]
penjualan_res = penjualan.reshape(-1, 1)

# Training model[cite: 1]
linreg = LinearRegression()
linreg.fit(penjualan_res, harga)

# Visualisasi hasil regresi[cite: 1]
plt.scatter(penjualan_res, harga, color='red')
plt.plot(penjualan_res, linreg.predict(penjualan_res))
plt.title("Visualisasi model regresi data penjualan dan harga")
plt.xlabel("Penjualan")
plt.ylabel("Harga")
plt.show()
```
